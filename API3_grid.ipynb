# Demo 3: Grid
- One important feature of KANs is that they embed splines to neural networks. 
- However, splines are only valid for approximating functions in known bounded regions, while the range of activations in neural networks may be changing over training. 
- So we have to update grids properly according to that. Let's first take a look at how we parametrize splines.

```python
from kan.spline import B_batch
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Consider a 1D example
# Suppose we have grid [-1, 1] with G intervals, spline order k
G = 5
k = 3
grid = torch.linspace(-1, 1, steps=G+1)[None, :]

# and we have sample range in [-1, 1]
x = torch.linspace(-1, 1, steps=1001)[None, :]

basis = B_batch(x, grid, k=k)
```

```python
for i in range(G+k):
    plt.plot(x[0].detach().numpy(), basis[0, i, :].detach().numpy())

plt.legend([f"B_{i}" for i in np.arange(G+k)])
plt.xlabel('x')
plt.ylabel(f'B_i(x)')
```
